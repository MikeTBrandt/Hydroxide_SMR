```python
%matplotlib inline
from math import pi, sin, cos
import numpy as np
import openmc
import openmc.deplete
from matplotlib import pyplot
```

```python
results1 = openmc.deplete.Results("./depletion_results-1.h5")
time, k1 =results1.get_keff()

results2 = openmc.deplete.Results("./depletion_results-2.h5")
time, k2 =results2.get_keff()

results3 = openmc.deplete.Results("./depletion_results-3.h5")
time, k3 =results3.get_keff()

results4 = openmc.deplete.Results("./depletion_results-4.h5")
time, k4 =results4.get_keff()

results5 = openmc.deplete.Results("./depletion_results-5.h5")
time, k5 =results5.get_keff()
```

```python
time /=(24*3600*1201.75)/42.06 #converts to MWd/kgHM
eff=[1]*201
```

```python
pyplot.plot(time, k1[:, 0],color='orange',marker='o',markersize=3, label='Lat2-ModG-Cool4')
pyplot.plot(time, k2[:, 0],color='green',marker='o',markersize=3,  label='Lat2-Mod7LiOH-Cool4')
pyplot.plot(time, k3[:, 0],color='black',marker='o',markersize=3,  label='Lat2-ModZrOD-Cool4')
pyplot.plot(time, k4[:, 0],color='pink',marker='o',markersize=3,   label='Lat2-Mod7LiOH-Cool7')
pyplot.plot(time, k5[:, 0],color='purple',marker='o',markersize=3, label='Lat2-ModG-Cool7')
pyplot.plot(time,eff,color='red')        
pyplot.xlabel("Burnup (MWd/kg)")
pyplot.ylabel("$k_{eff}$");
pyplot.xlim(0,160)
pyplot.ylim(.8,1.5)
pyplot.grid()
pyplot.legend(loc='upper right')
pyplot.savefig("Lattice_3_Results.png",dpi=300)
pyplot.show()
```
