```python
%matplotlib inline
import openmc
```

```python
u=openmc.Material(1,"u");
u.add_element('U',1.0,enrichment=5);

water=openmc.Material(2,"h2o");
water.add_nuclide('H1',2.0);
water.add_nuclide('O16',1);
water.set_density('g/cm3',1.0);
water.add_s_alpha_beta('c_H_in_H2O');

materials = openmc.Materials([u,water]);
materials.export_to_xml();
```

```python
Fuel_Sphere=openmc.Sphere(r=5);
Moderator_Sphere=openmc.Sphere(r=25,boundary_type='vacuum');


Fuel_Cell=openmc.Cell();
Fuel_Cell.fill=u;
Fuel_Cell.region=-Fuel_Sphere;

Moderator_Cell=openmc.Cell();
Moderator_Cell.fill=water;
Moderator_Cell.region=+Fuel_Sphere & -Moderator_Sphere

root_universe=openmc.Universe();
root_universe.add_cells([Fuel_Cell,Moderator_Cell]);

geometry = openmc.Geometry(root_universe);
geometry.export_to_xml()
```

```python
settings = openmc.Settings()

bounds = [-5,-5,-5,5,5,5];
uniform_dist = openmc.stats.Box(bounds[:3],bounds[3:],
                                only_fissionable=True);
settings.source=openmc.source.Source(space=uniform_dist);
settings.batches = 1000
settings.inactive = 10
settings.particles = 1000
settings.export_to_xml()
```
